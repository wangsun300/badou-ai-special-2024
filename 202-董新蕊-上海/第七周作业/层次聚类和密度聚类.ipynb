```python
import numpy as np
from scipy.cluster.hierarchy import dendrogram, linkage,fcluster
from matplotlib import pyplot as plt
```

```python
x = np.array([[1,2],[5,2],[7,9],[6,0],[6,9],[4,2],[5,7]])

"""
运用spicy中的linkage对其进行层次聚类计算。linkage(输入要计算的数组，method=计算距离的方法)
方法包括：single单链法（容易形成链式结果），complete全链法, ward计算最小方差（常用）
"""
#选择ward方式计算距离效果更好
y = linkage(x, method = 'ward')
print(y)
max_d = 3
clusters = fcluster(y, max_d, criterion='distance')
print(clusters)

#将层次聚类编码转为树状图的链接矩阵，返回一个字典，包含绘制树状图的信息。

dendrogram(y) 
#运用matplotlib绘制树状图
#plt.figure(figsize=(10,7))
#plt.xlabel('Sample index')
#plt.ylabel('Distance')
#plt.title('Hierarchy Cluster')


# 展示离散点
plt.figure(figsize=(8,6))
plt.scatter(x[:,0],x[:,1], c=clusters, cmap = 'viridis')
plt.title('Hierarachical Clustering Result')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show
```
