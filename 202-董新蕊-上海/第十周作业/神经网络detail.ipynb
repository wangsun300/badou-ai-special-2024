```python
import scipy.special
import numpy as np

class NeauralNetwork:
    def __init__(self, inputnodes, hiddennodes, outputnodes, learningrate):
        #网络结构初始化，输入层中间层和输出层节点个数
        self.inodes = inputnodes
        self.hnodes = hiddennodes
        self.onodes = outputnodes
        
        #设置学习率
        self.lr = learningrate
        pass
        
        #设置初始化权重 wih为input和hidden层的权重，who为hidden和output层的权重
        #numpy.random.rand()函数输入的是数组维度，返回的是0-1之间的随机数
        self.wih = np.random.rand(self.hnodes, self.inodes) - 0.5
        self.who = np.random.rand(self.onodes, self.hnodes) - 0.5
        
        #设置激活函数 scipy.special.expit对应的是sigmoid函数
        #lambda是python的关键字，类似C语言中的宏定义，当调用self.activation_function(x）时，会自动转换为之后的函数
        self.activation_function = lambda x:scipy.special.expit(x)
        
    def train(self, inputs_list, targets_list):
        '''
        1. 正向传播
        将inputs和outputs转化为numpy支持的二维矩阵
    
        '''
        inputs = np.array(inputs_list, ndmin = 2).T
        targets = np.array(targets_list, ndmin = 2).T
    
        hidden_inputs = np.dot(self.wih, inputs)
        hidden_outputs = self.activation_function(hidden_inputs)
        final_inputs = np.dot(self.who, hidden_outputs)
        final_outputs = self.activation_function(final_inputs)

        '''
        2. 反向传播
        对比输出值和目标值，计算误差，计算梯度，重新更新权重参数
        
        '''
        outputs_errors = targets - final_outputs
        hidden_errors = np.dot(self.who.T, outputs_errors * final_outputs * (1 - final_outputs)) 
    
        self.who += self.lr * np.dot((outputs_errors * final_outputs * (1 - final_outputs)), hidden_outputs.T)
        self.wih += self.lr * np.dot((hidden_errors * hidden_outputs * (1 - hidden_outputs)), inputs.T)
    
    def query(self, inputs_list):
        
        inputs = np.array(inputs_list, ndmin=2).T
        
        hidden_inputs = np.dot(self.wih, inputs)
        hidden_outputs = self.activation_function(hidden_inputs)
        final_inputs = np.dot(self.who, hidden_outputs)
        final_outputs = self.activation_function(final_inputs)
        
        print(final_outputs)
        return final_outputs
```

```python
#定义网络结构具体内容  图像为28*28 所以是784个输入节点
input_nodes = 784
hidden_nodes = 100
output_nodes = 10
learaning_rate = 0.1
n = NeauralNetwork(input_nodes, hidden_nodes, output_nodes, learaning_rate)

#加入训练数据
training_data_file = open("dataset/mnist_train.csv", 'r')
training_data_list = training_data_file.readlines()
training_data_file.close()

#加入eopchs
epochs = 10
for i in range(epochs):
    for record in training_data_list:
        all_values = record.split(',')
        inputs = (np.asfarray(all_values[1:])) / 255.0 * 0.99 + 0.01
        targets = np.zeros(output_nodes) + 0.01
        targets[int(all_values[0])] = 0.99    
        n.train(inputs, targets)
```

```python
'''
对训练好的模型开始测试
'''
#加入测试数据
test_data_file = open("dataset/mnist_test.csv", 'r')
test_data_list = test_data_file.readlines()
test_data_file.close()

scores = []

for record in test_data_list:
    all_values = record.split(',')
    inputs = (np.asfarray(all_values[1:])) / 255.0 * 0.99 + 0.01
    #图像预测
    prediction_outputs= n.query(inputs)  

    #和正确值对比
    
    correct_number = int(all_values[0])
    print('correct number', correct_number)
    
    label = np.argmax(prediction_outputs)
    print('prediction number', label) 
    if label == correct_number:
        scores.append(1)
    else:
        scores.append(0)
print(scores)

scores_array = np.asarray(scores)
print('performance = ', scores_array.sum() / scores_array.size)
```
