```python
import numpy as np
import matplotlib.pyplot as plt
 
def Normalization1(x):
    #放缩到0-1之间
    # x=(x-min(x))/max(x)-min(x)
    return [(float(i) - min(x))/max(x) - min(x) for i in x]
def Normalization2(x):
    #放缩到-1-1之间
    # x=(x-mean(x))/max(x)-min(x)
    return [(float(i) - np.mean(x))/max(x) - min(x) for i in x]

def z_score(x):
    #zero-mean normalization
    #将数据按照行或列或其他属性值减去其均值再除以其标准差
    #符合正态分布，均值为0，标准差为1
    std =sum([(i-np.mean(x))*(i-np.mean(x)) for i in x])/len(x)
    return [(float(i) - np.mean(x))/std for i in x]

l=[-10, 5, 5, 6, 6, 6, 7, 7, 7, 7, 8, 8, 8, 8, 8, 9, 9, 9, 9, 9, 9, 10, 10, 10, 10, 10, 10, 10, 11, 11, 11, 11, 11, 11, 12, 12, 12, 12, 12, 13, 13, 13, 13, 14, 14, 14, 15, 15, 30]


n=Normalization2(l)
z=z_score(l)

cs = []
for i in l:
    c=l.count(i)
    cs.append(c)
print(cs)

plt.plot(l,cs) #原始数据分布
plt.plot(z,cs) #标准化后数据分布
plt.plot(n,cs) #归一化后数据分布
plt.show()
```
